# 4. Select samples for the following analysis

```python
meta = project.get_meta_table(project="PDAC-105")
meta
```

```python
# read sample names
tumor_samples = []
nat_samples = []
nd_samples = []

for index,row in meta.iterrows():
    status = row['Pathological.Status']
    sample = row['Sample.ID']
    if status == "Tumor":
        tumor_samples.append(sample)
    elif status == 'NAT':
        nat_samples.append(sample)
    elif status == "Normal-duct":
        nd_samples.append(sample)
    else:
        print(sample)

len(tumor_samples),len(nat_samples),len(nd_samples)
```

```python
# get all samples from the oringal data 
data = glycoproteomics[tumor_samples+nat_samples+nd_samples]
data
```

```python
from gpnotebook.tools.clinical import get_simplified_stage
# simplify stage types
stage_map = dict(clinical['tumor_stage_pathological'])
meta['Stage_Curated'] = [get_simplified_stage(stage_map.get(i)) for i in meta['Patient.ID']]
# build map of sample to stage
stage_d = dict(zip(meta['Sample.ID'],meta['Stage_Curated']))

early_stage_samples = [i for i in tumor_samples if stage_d.get(i) in ['I','II']]
len(early_stage_samples)
```

# 6. Statistical tests

```python
# differential expression analysis
from omicsone.plugins.diff import compare_two_groups

# calculate statistic test
diff1 = compare_two_groups(data,early_stage_samples,nat_samples,method="Wilcoxon(Unpaired)",
                           max_miss_ratio_global=0.5, max_miss_ratio_group=0.5,fdr_cutoff=0.01, log2fc_cutoff=1)
diff2 = compare_two_groups(data,early_stage_samples,nd_samples,method="Wilcoxon(Unpaired)",
                           max_miss_ratio_global=0.5, max_miss_ratio_group=0.5,fdr_cutoff=0.01, log2fc_cutoff=1)
```

```python
def find_markers(diff1,diff2):

    a = set(diff1[diff1['Significance']=='S-U'].index)
    b = set(diff2[diff2['Significance']=='S-U'].index)

    x = a & b
    
    return x

es_markers = find_markers(diff1,diff2)
print(len(es_markers))
```

```python
es_markers_genes = set([i[0].split("|")[0] for i in es_markers])
len(es_markers_genes)
```

```python
from gpnotebook.tools.utils import get_cc_database
cc_database = get_cc_database()
cc_database
```

```python
rows = []
for index,row in cc_database.iterrows():
    if pd.isna(row['Gene Names']):
        continue
    genes = set(re.split('\s',row['Gene Names']))
    genes = genes.intersection(es_markers_genes)
    if len(genes) == 0:
        continue
    gene = sorted(list(genes))[0]
    scc = row['Subcellular location [CC]']
    gocc = row['Gene Ontology (cellular component)']
    status = 'NA'
    if pd.notna(scc) and re.search('Secreted',scc,re.IGNORECASE):
        status = 'Secreted'
        rows.append([gene,status])
        continue
    elif pd.notna(gocc) and re.search('Extracellular Space',gocc, re.IGNORECASE):
        status = 'Extracellular space'
        rows.append([gene,status])
        continue
    elif pd.notna(gocc) and re.search('Extracellular Region',gocc, re.IGNORECASE):
        status = 'Extracellular region'
        rows.append([gene,status])
        continue
    elif pd.notna(gocc) and re.search('extracellular', gocc,re.IGNORECASE):
        status = 'Extracellular (other)'
        rows.append([gene,status])
        continue
    elif pd.notna(gocc) and re.search('surface|membrane',gocc,re.IGNORECASE):
        status = 'Surface/Membrane'
        rows.append([gene,status])
        continue
    elif pd.notna(gocc) and re.search('endoplasmic reticulum',gocc,re.IGNORECASE):
        status = 'ER'
        rows.append([gene,status])
        continue
    else:
        print(gene,scc,gocc)
        
cc_df = pd.DataFrame(rows,columns=['Gene','Cellular Location']).drop_duplicates()
cc_df
```
