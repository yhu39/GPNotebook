# 4. Select samples for the following analysis

```python
meta = project.get_meta_table(project="PDAC-105")
meta
```

```python
# read sample names
tumor_samples = []
nat_samples = []
nd_samples = []

for index,row in meta.iterrows():
    status = row['Pathological.Status']
    sample = row['Sample.ID']
    if status == "Tumor":
        tumor_samples.append(sample)
    elif status == 'NAT':
        nat_samples.append(sample)
    elif status == "Normal-duct":
        nd_samples.append(sample)
    else:
        print(sample)

len(tumor_samples),len(nat_samples),len(nd_samples)
```

```python
# get all samples from the oringal data 
data = glycoproteomics[tumor_samples+nat_samples+nd_samples]
data
```

# 6. Statistical tests

```python
# differential expression analysis
from omicsone.plugins.diff import compare_two_groups

# calculate statistic test
diff1 = compare_two_groups(data,tumor_samples,nat_samples,method="Wilcoxon(Unpaired)",
                           max_miss_ratio_global=0.5, max_miss_ratio_group=0.5,fdr_cutoff=0.01, log2fc_cutoff=1)
diff2 = compare_two_groups(data,tumor_samples,nd_samples,method="Wilcoxon(Unpaired)",
                           max_miss_ratio_global=0.5, max_miss_ratio_group=0.5,fdr_cutoff=0.01, log2fc_cutoff=1)
```

```python
diff1
```

```python
genes = []
degps = list(diff1[diff1['Significance']=='S-U'].index) + list(diff1[diff1['Significance']=='S-D'].index)
print(len(degps))
for i in degps:
    genes.append(i[0])
print(len(set(genes)))
```

```python
diff2
```

```python
# save results
diff1_tsv = os.path.join(job_dir,'Glycopeptide_Tumor_NAT_diff.tsv')
diff2_tsv = os.path.join(job_dir,'Glycopeptide_Tumor_NormalDuct_diff.tsv')

diff1_joblib = os.path.join(job_dir,'Glycopeptide_Tumor_NAT_diff.joblib')
diff2_joblib = os.path.join(job_dir,'Glycopeptide_Tumor_NormalDuct_diff.joblib')

import joblib

diff1.to_csv(diff1_tsv,sep="\t")
diff2.to_csv(diff2_tsv,sep="\t")

joblib.dump(diff1, diff1_joblib)
joblib.dump(diff2, diff2_joblib)
```

# 8. Enrichment analysis

```python
# import dependencies

from gpnotebook.tools.enrichment import enrichr
```

```python
total_genes = set([i[0].split('|')[0] for i in data.index])
up_genes = set([i[0].split('|')[0] for i in diff1[diff1['Significance']=='S-U'].index])
down_genes = set([i[0].split('|')[0] for i in diff1[diff1['Significance']=='S-D'].index])
pure_down_genes = down_genes - up_genes
pure_up_genes = up_genes - down_genes
len(total_genes),len(up_genes),len(down_genes),len(pure_down_genes),len(pure_up_genes)
```

```python
import time
up_enrichr_df = enrichr(pure_up_genes,total_genes,job_dir)
time.sleep(5)
down_enrichr_df = enrichr(pure_down_genes,total_genes,job_dir)
```

```python
# plot enriched pathways
skip_pathways = [
    'Phagosome','Human papillomavirus infection','Pertussis',
    'Malaria','Arrhythmogenic right ventricular cardiomyopathy',
    'Staphylococcus aureus infection','Regulation of actin cytoskeleton'
]
up_enrichr_df2 = up_enrichr_df[~up_enrichr_df['Term'].isin(skip_pathways)].sort_values('Adjusted P-value').head(10)
down_enrichr_df2 = down_enrichr_df[~down_enrichr_df['Term'].isin(skip_pathways)].sort_values('Adjusted P-value').head(10)

from gpnotebook.tools.visualization import plot_enrichr_both
fig = plot_enrichr_both(up_enrichr_df2,down_enrichr_df2,title='KEGG Pathways Enrichment Analysis',min_x=-30,max_x=30)
```

# 9. Stage difference

```python
from gpnotebook.tools.clinical import get_simplified_stage
# simplify stage types
stage_map = dict(clinical['tumor_stage_pathological'])
meta['Stage_Curated'] = [get_simplified_stage(stage_map.get(i)) for i in meta['Patient.ID']]
# build map of sample to stage
stage_d = dict(zip(meta['Sample.ID'],meta['Stage_Curated']))

early_stage_samples = [i for i in tumor_samples if stage_d.get(i) in ['I','II']]
len(early_stage_samples)
```

```python
import ipywidgets as widgets
from IPython.display import display
from gpnotebook.tools.visualization import boxplot_groups


# Options for the dropdown
gp_list = list(diff1.index)
options = ['@'.join([str(k) for k in i]) for i in gp_list]
sorted_options = sorted(options)
options_map = dict(zip(options,gp_list))

groups = {
    '2.Tumor': tumor_samples,
    '1.NAT': nat_samples,
    "3.EarlyStage": early_stage_samples
}

dropdown = widgets.Dropdown(
    options=sorted_options,
    value=sorted_options[0],
    description='Choose:',
    disabled=False,
)

output = widgets.Output()

# Event handler for the dropdown change
def on_dropdown_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        with output:
            gp = options_map.get(change['new'])
            boxplot_groups(data,groups, gp)

    
    
# Display the dropdown, button, and output area
display(dropdown, output)

# Attach the event handlers
dropdown.observe(on_dropdown_change)
```
