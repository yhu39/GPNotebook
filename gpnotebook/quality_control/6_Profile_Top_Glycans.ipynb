```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

from gpnotebook.tools.standard_imports import *
```

```python
cancers = cptac_glyco.get_cancer_info()
```

```python
cancers
```

```python
counters = []

for cancer in cancers:
    cancer_glycan_counter = cptac_glyco.get_cancer(cancer).count_glycans()
    counters.append(cancer_glycan_counter)

df = pd.DataFrame(counters,index = cancers)
```

```python
df.fillna(0,inplace=True)
```

```python
df
```

```python
# Calculating row sums
row_sums = df.sum(axis=1)

# Converting counts to ratios
df_ratio = df.div(row_sums, axis=0)
df_ratio
```

```python
#%%
top_glycans = sorted(dict(df.sum(axis=0)).items(), key = lambda i: i[1],reverse=True)[:10]
top_glycans
```

```python
top_glycans_df = df_ratio.loc[:,[i[0] for i in top_glycans]]
top_glycans_df.index = [str(i).upper() for i in top_glycans_df.index]
top_glycans_df
```

```python
# out_fig_path = outdir + os.sep + "Figure2_TopGlycans_20230123.png"
fig,ax = plt.subplots(figsize=(6,4))
# top_df.plot(kind="bar",stacked=True,ax=ax,color=sns.color_palette("husl", 13))
top_glycans_df.plot(kind="bar",stacked=True,ax=ax,color=sns.color_palette("Paired",11))
plt.title("Occupations of top 10 glycans on glycosites")
plt.xlabel("Cohorts")
plt.ylabel("Percentages")
# ax.set_yticklabels(['0%','20%','40%','60%','80%','100%'])
ax.set_yticklabels(['0%','5%','10%','15%','20%','25%','30%'])
plt.legend(bbox_to_anchor=(1.35, 1),loc='upper right')
# plt.savefig(out_fig_path,bbox_inches='tight')
```
