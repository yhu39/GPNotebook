```python
fig,ax = plt.subplots(figsize=(6,4))
n = 0

enrichment_colors = dict(zip(['G','P','P+G'],['r','g','b']))
enrichments = []

for index,row in df.iterrows():
    em = row['Enrichment.Method']
    color = enrichment_colors.get(em)
    if em not in enrichments:
        plt.bar([n],[row['Glycopeptides']],color=color,label='Enrichment: {}'.format(row['Enrichment.Method']))
        enrichments.append(em)
    else:
        plt.bar([n],[row['Glycopeptides']],color=color)
    n += 1
sns.lineplot(data=df, x=df.index, y="Cumulative.Glycopeptides",
             label='Cumulative identified glycopeptides',color='black')
t = plt.title("Identified Glycopeptides")
plt.xlabel('Cancer Type')
# fig.savefig(figure1b_path,bbox_inches='tight')
```
