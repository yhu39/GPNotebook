```python
cancers = cptac_glyco.get_cancer_info()
cancers
```

```python
glycoproteins = []
glycopeptides = []

for cancer in cancers:
    cancer_glyco = cptac_glyco.get_cancer(cancer).get_glycoproteomics()
    for index,row in cancer_glyco.iterrows():
        glycopeptide = '@'.join([str(i) for i in index])
        glycoprotein = index[0]
        glycopeptides.append(glycopeptide)
        glycoproteins.append(glycoprotein)

print(len(set(glycopeptides)),len(set(glycoproteins)))
```
