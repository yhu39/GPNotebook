```python
# read sample names
tumor_samples = []
nat_samples = []
nd_samples = []

for index,row in meta.iterrows():
    status = row['Pathological.Status']
    sample = row['Sample.ID']
    if status == "Tumor":
        tumor_samples.append(sample)
    elif status == 'NAT':
        nat_samples.append(sample)
    elif status == "Normal-duct":
        nd_samples.append(sample)
    else:
        print(sample)

len(tumor_samples),len(nat_samples),len(nd_samples)
```

```python
samples_in_gyco = tumor_samples + nat_samples
pdac_glyco = pdac_glyco[samples_in_gyco]
```

```python
pdac_phospho = cptac.Pdac().get_phosphoproteomics('umich').T
pdac_phospho.columns = ['PDA.' + i if re.search('\.[NT]',i) else 'PDA.' + i +  ".T" for i in pdac_phospho.columns.values ]
pdac_phospho.head(2)
```

```python
# differential expression analysis
from omicsone.plugins.diff import compare_two_groups

# calculate statistic test
diff_glyco = compare_two_groups(pdac_glyco,tumor_samples,nat_samples,method="Wilcoxon(Unpaired)",
                           max_miss_ratio_global=0.5, max_miss_ratio_group=0.5,fdr_cutoff=0.01, log2fc_cutoff=1)
diff_phospho = compare_two_groups(pdac_phospho,tumor_samples,nat_samples,method="Wilcoxon(Unpaired)",
                           max_miss_ratio_global=0.5, max_miss_ratio_group=0.5,fdr_cutoff=0.01, log2fc_cutoff=1)
```

```python
diff_glyco.head(2)
```

```python
diff_phospho.head(2)
```

```python
glyco_map = dict()
for index,row in tqdm(diff_glyco.iterrows()):
    gene = index[0].split('|')[0]
    if gene not in gene_list:
        continue
    if gene not in glyco_map:
        glyco_map[gene] = []
    feature = '@'.join([str(i) for i in index])
    glyco_map[gene].append((feature, row['Log2FC(median)'],row['Significance']))
```

```python
phospho_map = dict()
for index,row in tqdm(diff_phospho.iterrows()):
    gene = index[0]
    if gene not in gene_list:
        continue
    if gene not in phospho_map:
        phospho_map[gene] = []
    feature = '@'.join([str(i) for i in index])
    phospho_map[gene].append((feature, row['Log2FC(median)'],row['Significance']))
```

```python
rows = []
for gene in gene_list:
    if gene not in glyco_map or gene not in phospho_map:
        continue
    
    for g_feature,g_log2FC,g_sign in glyco_map[gene]:
        for p_feature,p_log2FC,p_sign in phospho_map[gene]:
            sign = 'N/N'
            if g_sign == 'S-U':
                g = 'U'
            elif g_sign == 'S-D':
                g = 'D'
            else:
                g = 'N'

            if p_sign == 'S-U':
                p = 'U'
            elif p_sign == 'S-D':
                p= 'D'
            else:
                p = 'N'
    
            sign = f'{g}/{p}'
                
            rows.append([gene,g_feature,g_log2FC,p_feature,p_log2FC,sign])
result = pd.DataFrame(rows,columns=['Gene','G.Feature','G.Log2FC','P.Feature','P.Log2FC','Sign(G/P)'])
```

```python
result = result.sort_values('Sign(G/P)')
```

```python
from scipy.stats import linregress
import numpy as np

# Create two arrays of data
# x = np.array([1, 2, 3, 4, 5])
# y = np.array([2, 4, 6, 8, 10])

x = result['G.Log2FC']
y = result['P.Log2FC']

# Perform linear regression
a, b, r_value, p_value, std_err = linregress(x, y)

print("Slope: ", a)
print("Intercept: ", b)
print("p_value",p_value)
```
