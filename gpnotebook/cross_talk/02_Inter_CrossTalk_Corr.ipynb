```python
# read sample names
tumor_samples = []
nat_samples = []
nd_samples = []

for index,row in meta.iterrows():
    status = row['Pathological.Status']
    sample = row['Sample.ID']
    if status == "Tumor":
        tumor_samples.append(sample)
    elif status == 'NAT':
        nat_samples.append(sample)
    elif status == "Normal-duct":
        nd_samples.append(sample)
    else:
        print(sample)

len(tumor_samples),len(nat_samples),len(nd_samples)
```

```python
samples_in_gyco = tumor_samples + nat_samples
pdac_glyco = pdac_glyco[samples_in_gyco]
```

```python
pdac_phospho = cptac.Pdac().get_phosphoproteomics('umich').T
pdac_phospho.columns = ['PDA.' + i if re.search('\.[NT]',i) else 'PDA.' + i +  ".T" for i in pdac_phospho.columns.values ]
pdac_phospho.head(2)
```

```python
samples_in_phospho = pdac_phospho.columns.values
common_samples = sorted(list(set(samples_in_gyco).intersection(set(samples_in_phospho))))
len(common_samples)
```

```python
pdac_glyco = pdac_glyco[common_samples]
pdac_phospho = pdac_phospho[common_samples]
```

```python
pdac_glyco.shape,pdac_phospho.shape
```

```python
com_tumor_samples = [i for i in common_samples if i in tumor_samples]
com_nat_samples = [i for i in common_samples if i in nat_samples]
print(len(com_tumor_samples),len(com_nat_samples))

# differential expression analysis
from omicsone.plugins.diff import compare_two_groups

# calculate statistic test
diff_glyco = compare_two_groups(pdac_glyco,com_tumor_samples,com_nat_samples,method="Wilcoxon(Unpaired)",
                           max_miss_ratio_global=0.5, max_miss_ratio_group=0.5,fdr_cutoff=0.01, log2fc_cutoff=1)
diff_phospho = compare_two_groups(pdac_phospho,com_tumor_samples,com_nat_samples,method="Wilcoxon(Unpaired)",
                           max_miss_ratio_global=0.5, max_miss_ratio_group=0.5,fdr_cutoff=0.01, log2fc_cutoff=1)
```

```python
pdac_tumors_glyco = pdac_glyco[com_tumor_samples]
pdac_tumors_phospho = pdac_phospho[com_tumor_samples]
```

```python
rows = []
for index,row in pdac_tumors_glyco.iterrows():
    missin_values_count = len([i for i in row if pd.isna(i)])
    miss_ratio = missin_values_count / len(common_samples)
    if miss_ratio < 0.5:
        rows.append(row)
pdac_glyco = pd.DataFrame(rows)
pdac_glyco.shape
```

```python
rows = []
for index,row in pdac_tumors_phospho.iterrows():
    missin_values_count = len([i for i in row if pd.isna(i)])
    miss_ratio = missin_values_count / len(common_samples)
    if miss_ratio < 0.5:
        rows.append(row)
pdac_phospho = pd.DataFrame(rows)
pdac_phospho.shape
```

```python
pdac_phospho_map = dict()
for index,row in pdac_phospho.dropna().iterrows():
    key = index
    value = list(row)
    pdac_phospho_map[key] = value

pdac_glyco_map = dict()
for index,row in pdac_glyco.dropna().iterrows():
    key = index
    value = list(row)
    pdac_glyco_map[key] = value
```

```python
from scipy.stats import spearmanr
rows = []
for i in tqdm(pdac_glyco_map):
    x = pdac_glyco_map[i]
    for j in pdac_phospho_map:
        y = pdac_phospho_map[j]
        corr,pval = spearmanr(x,y)
        rows.append([i,j,corr,pval])
        
```

```python
corr_df = pd.DataFrame(rows,columns=['GP','PP','Corr','P-Value'])
```

```python
import requests

string_api_url = "https://string-db.org/api"
output_format = "json"  # JSON format might be easier to parse
method = "get_string_ids"

params = {
    "identifiers": "\r".join(list(glyco_genes) + list(phospho_genes)),  # Concatenate both gene lists
    "species": 9606,  # Homo sapiens (human) NCBI identifier
    "limit": 1,  # only one (best) identifier per input protein
    "echo_query": 1,  # echoes the input identifiers in the output
    "caller_identity": "your_app_name"  # your app or your email
}

request_url = "/".join([string_api_url, output_format, method])

response = requests.post(request_url, data=params)

# Parsing JSON response
identifiers_response = response.json()

# You would need to extract STRING identifiers from the response.
# Depending on the structure of `identifiers_response`, the code to do this will vary.
```

```python
interaction_method = "network"

interaction_params = {
    "identifiers": "\r".join([item["stringId"] for item in identifiers_response]),  # STRING IDs retrieved from step 1
    "species": 9606,
    "network_flavor": "confidence",  # Use "actions" to get all interactions
    "caller_identity": "your_app_name"
}

interaction_request_url = "/".join([string_api_url, output_format, interaction_method])

interaction_response = requests.post(interaction_request_url, data=interaction_params)

# Again, you'll get a JSON response, which you'll need to parse to extract the interaction data.
```

```python
import json
interactions = interaction_response.json()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Assuming `interaction_response` is a list of edges between genes:
G = nx.Graph()

genes_in_graph = []
# Set default node color
default_color = 'green'  # You can change this to your preferred default color

for interaction in interactions:
    node_a = interaction["preferredName_A"]
    node_b = interaction["preferredName_B"]

    interaction_score = interaction['score']
    
    if (node_a in glyco_genes and node_b in glyco_genes) or (
        node_a in phospho_genes and node_b in phospho_genes):
        continue
    

    genes_in_graph.append(node_a)
    genes_in_graph.append(node_b)

    # if node_a in phospho_genes:
    #     node_a = node_a + "*"

    # if node_b in phospho_genes:
    #     node_b = node_b + "*"
    
    G.add_edge(node_a, node_b, weight=interaction_score)
    
    # Add or update attributes for node A
    if node_a in glyco_genes:
        G.nodes[node_a]['color'] = [0.3,0.3,1,0.5]
    elif node_a in phospho_genes:
        G.nodes[node_a]['color'] = [1,0.3,0.3,0.5]
    else:  # if node is not in s1_genes or s2_genes
        G.nodes[node_a].setdefault('color', default_color)
    
    # Add or update attributes for node B
    if node_b in glyco_genes:
        G.nodes[node_b]['color'] = [0.3,0.3,1,0.5]
    elif node_b in phospho_genes:
        G.nodes[node_b]['color'] = [1,0.3,0.3,0.5]
    else:  # if node is not in s1_genes or s2_genes
        G.nodes[node_b].setdefault('color', default_color)

# Get a list of colors for each node
colors = [G.nodes[node].get('color', default_color) for node in G]

# Draw the graph
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G, seed=42)  # Using a fixed seed for reproducible layout
# pos = nx.circular_layout(G)
# pos = nx.kamada_kawai_layout(G)
# pos = nx.spectral_layout(G)
# pos = nx.shell_layout(G)
# pos = nx.random_layout(G)





# Draw nodes with their corresponding colors
nx.draw_networkx_nodes(G, pos, node_color=colors)

# Draw edges and labels
nx.draw_networkx_edges(G, pos, alpha=0.5)
nx.draw_networkx_labels(G, pos)

# Show the plot
plt.axis('off')  # Turn off the axis
plt.show()
```

```python
pathway_color_map = {
    'Epithelial Mesenchymal Transition': [1,0,0,0.5],
    'Coagulation': [0,1,0,0.5],
    'Myogenesis': [0,0,1,0.5],
    "Angiogenesis": [1,0,1,0.5]
}

nodes = {}
pathways = dict()
for index,row in enr.results.iloc[:5].iterrows():
    term = row['Term']
    genes = row['Genes'].split(';')
    if term not in pathway_color_map:
        continue
    pathways[term] = genes

    for gene in genes:
        if gene not in nodes:
            nodes[gene] = []
        nodes[gene].append(term)
```
